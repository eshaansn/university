```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
import sklearn.decomposition
```

```python
df = pd.read_csv("data/winedata.csv")
```

```python
x = df.drop('Class', axis=1)
y = df['Class']
```

```python
x = StandardScaler().fit_transform(x)
x = pd.DataFrame(data=x, columns= df.columns[1:])
x.head()
```

```python
pca = sklearn.decomposition.PCA(n_components=None)
pcax = pca.fit_transform(x)
pcax = pd.DataFrame(data=pcax)
```

```python
plt.figure(figsize = (11, 6))
plt.scatter(pcax[0], pcax[1], c=y, alpha=0.5, s=148, edgecolors='black')
```

```python
from sklearn.manifold import TSNE
import seaborn as sns
```

```python
tsne = TSNE(n_components=2, n_iter=1000, perplexity=100)
tsne_x = tsne.fit_transform(x)
```

```python
tsne_x = tsne_x.T
sns.scatterplot(x=tsne_x[0], y=tsne_x[1], hue=y)
```
