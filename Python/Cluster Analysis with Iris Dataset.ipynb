```python
from sklearn import datasets
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
iris = datasets.load_iris()
features = iris.data
labels = iris.target
iris_df = pd.DataFrame(data = features, columns=['Sepal_Length', 'Sepal_Width', 'Petal_Length', 'Petal_Width'])
iris_df['Class'] = labels
```

```python
kmeans = KMeans(n_clusters=3)
prediction = kmeans.fit_predict(features)
iris_df['Prediction'] = prediction
```

```python
sns.scatterplot(x=iris_df['Sepal_Length'], y=iris_df['Sepal_Width'], hue=labels)
```

```python
sns.scatterplot(x=iris_df['Sepal_Length'], y=iris_df['Sepal_Width'], hue=prediction)
```

```python
kmeans.score(features, labels)
```
